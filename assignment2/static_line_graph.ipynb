```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio

data = pd.read_csv('sa_assign2.csv')

data['Date'] = pd.to_datetime(data['Date'])

for column in data.columns[1:]:
    data[column] = data[column].str.replace('%', '').astype(float)


fig = px.line(data, 
              x='Date', 
              y=data.columns[1:], 
              labels={'value': 'Cumulative Return (%)', 'variable': 'Strategy'}, 
              title='<b>Strategy Showdown: Alpha Picks vs. S&P 500’s Magnificent 7 and the Broader Market</b>'
              )

fig.update_layout(
    xaxis_title=dict(
        text='<b>Date</b>',
        font=dict(size=14)  
    ),
    yaxis_title=dict(
        text='<b>Cumulative Return (%)</b>',
        font=dict(size=14)  
    ),
    legend_title='Strategy',
    xaxis=dict(domain=[0, 0.98]),
    yaxis=dict(ticksuffix='%'),
    template='plotly',
    title=dict(
        text='<b>Strategy Showdown: Alpha Picks vs. S&P 500’s Magnificent 7 and the Broader Market</b>',
        font=dict(size=20, color='black'),
        x=0.5,  
        xanchor='center'  
    ), 
    width=1300,
    height=600
)

for column in data.columns[1:]:
    fig.add_trace(go.Scatter(
        x=[data['Date'].iloc[-1]],  
        y=[data[column].iloc[-1]],  
        text=[f"{data[column].iloc[-1]:.2f}%"],  
        mode='text',
        textposition='middle right',
        showlegend=False,
        textfont=dict(color='black', size=12)
    ))


fig.write_image("static_line_graph_sa2.png")
fig.show()
```
